**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

```python
my_array = np.random.randint(0, 10, 18)
print(my_array)

my_array[(3 <= my_array) & (my_array <= 8)] *= -1
print(my_array)
```

**2.** Заменить максимальный элемент случайного массива на 0

```python
print(my_array, f'max value is: {my_array.max()}')
my_array[my_array == my_array.max()] = 0
print(my_array)
```

**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

```python
my_array = np.random.randint(0, 6, (3, 2))
print(f"my_array:\n{my_array}\n")

grid = np.meshgrid(*my_array, indexing='ij')

result = np.array(grid).T.reshape(-1, len(my_array))

print(f"result:\n{result}")

```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
A_array = np.random.randint(0, 4, (8, 3))
B_array = np.random.randint(0, 4, (2, 2))
print("A_array:\n", A_array)
print("\nB_array:\n", B_array)

B_unique = np.unique(B_array)
print("\nUnique values in B_array: ", B_unique)

for row in A_array:
    flag = 0
    for i in B_unique:
        if i not in row:
            flag = 1
            break
    if flag == 0:
        print(row)

result = np.array([row for row in A_array if all(i in row for i in B_unique)])
print("\nResult:\n", result)
        
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
my_matrix = np.random.randint(0, 4, (10, 3))
print("Original array:\n", my_matrix)
 
indices = [len(np.unique(row)) > 1 for row in my_matrix]
result = my_matrix[indices]
print("\nResult array:\n", result)
```

**6.** Дан двумерный массив. Удалить те строки, которые повторяются

```python
my_matrix = np.random.randint(0, 3, (8, 3))
print("my_matrix:\n", my_matrix)

indices = ['True']*len(my_matrix)
indices = set()
for i, row_i in enumerate(my_matrix):
    for j, row_j in enumerate(my_matrix[i + 1:], start = i + 1):
        if np.array_equal(row_j, row_i):
            indices.add(j)
            indices.add(i)

print('\nindices:', indices)
result = np.delete(my_matrix, list(indices), axis=0)
print('\nresult:\n', result)
```

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
mtr = np.random.randint(0, 4, (5, 4))
print(f"Matrix:\n{mtr}\n")
# NumPy
values = mtr[np.arange(min(mtr.shape)), np.arange(min(mtr.shape))]
values = values[values != 0]

print(f"values: {values}")
result = np.prod(values)
print(f"result: {result}\n")

# Python
mtr = mtr.tolist()
print(f"Matrix:\n{mtr}\n")

result = 1
for i in range(min(len(mtr), len(mtr[0]))):
    if mtr[i][i] != 0:
        result *= mtr[i][i]
print(f"result: {result}")
```

* __Задача 2__: Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество.  
  Например, для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) ответ True.

```python
x = np.random.randint(0, 2, 4)
y = np.random.randint(0, 2, 4)
print(f"x: {x}\ny: {y}")

# NumPy
print("Numpy result: ", np.all(np.array_equal(np.sort(x), np.sort(y))))

# Python

x = sorted(x.tolist())
y = sorted(y.tolist())

print("Result Python: ", x == y)
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит ноль. 
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ 5.

```python
vec = np.random.randint(0, 10, 18)
print("vec: ", vec)
# Numpy
values = vec[np.where(vec[:-1] == 0)[0] + 1]

print(values)
print("NumPy max: ", values.max()) if len(values != 0) else print("No elements on the right of 0.")

# Python
vec = vec.tolist()
print("vec: ", vec)

values = [vec[index + 1] for index, value in enumerate(vec) if value == 0 and index != len(vec) - 1]

print(values) 
print("Python max: ", max(values)) if len(values) != 0 else print("No elements on the right of 0.")
```
