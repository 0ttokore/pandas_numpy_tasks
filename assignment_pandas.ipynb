```python
import pandas as pd
```

```python
# загружаем датасет
data = pd.read_csv("./data/adult.data.csv")
data.head()
```

**1. Посчитайте, сколько мужчин и женщин (признак *sex*) представлено в этом датасете**

```python
number_of_men = data['sex'].value_counts()
number_of_men
```

**2. Каков средний возраст мужчин (признак *age*) по всему датасету?**

```python
men_mean_age = data[data['sex'] == 'Male']['age'].mean()
men_mean_age
```

**3. Какова доля граждан Соединенных Штатов (признак *native-country*)?**

```python
portion_USA = data[data['native-country'] == 'United-States'].shape[0]
portion_USA
```

**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

```python
agg_functions = ["mean", "std"]
result_more50 = data[data['salary'] == '>50K']['age'].agg(agg_functions)
result_less50 = data[data['salary'] == '<=50K']['age'].agg(agg_functions)

print(f"result_more50:\n{result_more50}\n")
print(f"result_less50:\n{result_less50}")
```

**6. Правда ли, что люди, которые получают больше 50k, имеют минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
edu = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
result = data[(data['salary'] == '>50K') & (data['education'].isin(edu))].shape[0] / data[(data['salary'] == '>50K')].shape[0] * 100
print(f"{result:.2f}% of people with salary more than 50K have higher education")
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Asian-Pac-Islander*.**

```python
result = data.groupby(['race', 'sex']).describe()
result
```

```python
max_age = result.loc[('Asian-Pac-Islander', 'Male'), ('age', 'max')]
print(f"\nmax age of men with race 'Asian-Pac-Islander' is : {max_age}")
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
marital_status = ['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']

result = data[(data['salary'] == '>50K') & (data['marital-status'].isin(marital_status))].shape[0] / data[(data['salary'] == '>50K')].shape[0] * 100
print(f"{result:.2f}% of people with salary more than 50K are married")
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_hours_per_week = data['hours-per-week'].max()
print(f"max_hours_per_week: {max_hours_per_week}")

people_max_hours_per_week = data[data['hours-per-week'] == 99].shape[0]
print(f"people_max_hours_per_week: {people_max_hours_per_week}")

result = data[(data['hours-per-week'] == 99) & (data['salary'] == '>50K')].shape[0]/people_max_hours_per_week * 100
print(f"result: {result:.2f} %")

```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
young = data[(data['age'] >= 16) & (data['age'] < 35)]
adult = data[(data['age'] >= 35) & (data['age'] < 70)]
retiree = data[(data['age'] >= 70) & (data['age'] < 100)]

def age_func(age):
    if 16 <= age < 35:
        return 'young'
    elif 35 <= age < 70:
        return 'adult'
    elif 70 <= age < 100:
        return 'retiree'
    
data['AgeGroup'] = data['age'].apply(age_func)
data
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
result = data[data['salary'] == '>50K'].groupby(['AgeGroup']).size().reset_index(name = 'Count')
result
```

```python
res = result.loc[result['Count'].idxmax()]
print(res['AgeGroup'])
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
groups = data.groupby('occupation').agg(
    count=('occupation', 'size'),
    average_age=('age', 'mean'),
    all_hours_more_than_five=('hours-per-week', lambda x: (x > 5).all())
).reset_index()

print(groups)

def filter_func(df):
    return df[(df['average_age'] <= 40) & (df['all_hours_more_than_five'])]

filtered_groups = filter_func(groups)

print(f"\n{filtered_groups}")
```
